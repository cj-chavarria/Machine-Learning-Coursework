```python
import numpy as np
import matplotlib.pyplot as plt
```

Example with $h_\theta(x) = \theta\cdot x=\theta_1x_1$

```python
m = 6
X =  np.linspace(0, 5, m)
y = X
```

```python
plt.figure(figsize=(5,5))

theta1 = 1
h =  theta1 * X

plt.plot(X, h, 'b-', linewidth=2)
plt.plot(X, y, 'ro', markersize=6)

plt.xlabel('X', fontsize=12)
plt.ylabel('X', fontsize=12)
plt.grid()
plt.legend(['h', 'y'])
plt.show()
```

```python
plt.figure(figsize=(5,5))

theta1 = 0.5
h =  theta1 * X

plt.plot(X, h, 'g--', linewidth=2)
plt.plot(X, y, 'ro', markersize=6)

plt.xlabel('X', fontsize=12)
plt.ylabel('X', fontsize=12)
plt.grid()
plt.legend(['h', 'y'])
plt.show()
```

```python
theta1 =  np.linspace(-3,5, 101)
h =  [theta*X for theta in theta1]
J = (1/m)*((h-y)**2).sum(axis=1) #MSE

plt.plot(theta1, J, 'b-', linewidth=2)
plt.xlabel(r'$\theta_1$', fontsize=12)
plt.ylabel(r'$J(\theta_1)$', fontsize=12)
plt.grid()
plt.show()
```

```python
theta1 =  np.linspace(-3,5, 101)
h =  [theta*X for theta in theta1]
J = (1/m)*(np.abs(h-y)).sum(axis=1) #MAE

plt.plot(theta1, J, 'b-', linewidth=2)
plt.xlabel(r'$\theta_1$', fontsize=12)
plt.ylabel(r'$J(\theta_1)$', fontsize=12)
plt.grid()
plt.show()
```

```python
theta1 =  np.linspace(-3,5, 101)
h =  [theta*X for theta in theta1]
J = np.sqrt((1/m)*((h-y)**2).sum(axis=1)) #RMSE

plt.plot(theta1, J, 'b-', linewidth=2)
plt.xlabel(r'$\theta_1$', fontsize=12)
plt.ylabel(r'$J(\theta_1)$', fontsize=12)
plt.grid()
plt.show()
```

Example with $h_\theta(x) = \theta\cdot x=\theta_0+\theta_1x_1$

```python
m = 10

X = np.linspace(-2, 2, m)

y = 2.5+5*X
```

```python
theta0 = np.linspace(-2.5, 7.55, 51)
theta1 = np.linspace(0,10,51)

h = [[th0+th1*X for th0 in theta0] for th1 in theta1]
J = (1/m)*((h-y)**2).sum(axis=2) #MSE

fig = plt.figure(figsize=(5,5))
ax = plt.axes(projection='3d')
th0, th1 = np.meshgrid(theta0, theta1)
surf = ax.plot_surface(th0, th1, J, cmap='coolwarm')
ax.set_xlabel(r'$\theta_0$', fontsize=12)
ax.set_ylabel(r'$J(\theta_1)$', fontsize=12)
#ax.set_zlabel(r'$J(\theta_0, \theta_1)$', fontsize=12)
ax.text2D(-0.125, 0.4, r'$J(\theta_0, \theta_1)$', transform=ax.transAxes, rotation=90,fontsize=12)

fig.colorbar(surf, shrink=0.5, aspect=5)
ax.view_init(20,45)
```

```python
plt.contour(th0, th1, J,50, cmap='coolwarm')
ax.set_xlabel(r'$\theta_0$', fontsize=12)
ax.set_ylabel(r'$J(\theta_1)$', fontsize=12)
plt.colorbar()
plt.show()
```

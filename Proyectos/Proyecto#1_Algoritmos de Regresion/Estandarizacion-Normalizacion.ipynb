```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler

sns.set_theme()
```

```python
m = 500
rng = np.random.default_rng()
X = 2 * rng.random((m,1))
y = 4 + X + rng.standard_normal((m,1))
```

```python
fig, ax = plt.subplots(figsize = (10,7))

sns.scatterplot(x=X.flatten(), y=y.flatten(), ax=ax)
plt.xlabel('$x_1$')
plt.ylabel('y')
plt.show()
```

### Implementacion propia

Normalizacion

```python
X_norm = (X - min(X)) / (max(X) - min(X))
y_norm = (y - min(y)) / (max(y) - min(y))
```

```python
fig, ax = plt.subplots(figsize = (10,7))

sns.scatterplot(x=X_norm.flatten(), y=y_norm.flatten(), ax=ax, color='tab:red', label='Datos normalizados')
plt.xlabel('$x_1$')
plt.ylabel('y')
plt.show()
```

Estandarizacion

```python
X_std = (X - np.mean(X)) / np.std(X)
y_std = (y - np.mean(y)) / np.std(y)
```

```python
fig, ax = plt.subplots(figsize = (10,7))

sns.scatterplot(x=X_std.flatten(), y=y_std.flatten(), ax=ax, color='tab:green', label='Datos estandarizados')
plt.xlabel('$x_1$')
plt.ylabel('y')
plt.show()
```

### Utilizando Scikit-Learn

Normalizacion

```python
normalizacion = MinMaxScaler()

X_norm_sk = normalizacion.fit_transform(X)
y_norm_sk = normalizacion.fit_transform(y)
```

```python
fig, ax = plt.subplots(figsize = (10,7))

sns.scatterplot(x=X_norm_sk.flatten(), y=y_norm_sk.flatten(), ax=ax, color='purple', label='Datos normalizados (Sk-Learn)')
plt.xlabel('$x_1$')
plt.ylabel('y')
plt.show()
```

Estandarizacion

```python
estandarizacion = StandardScaler()

X_std_sk = estandarizacion.fit_transform(X)
y_std_sk = estandarizacion.fit_transform(y)
```

```python
fig, ax = plt.subplots(figsize = (10,7))

sns.scatterplot(x=X_std_sk.flatten(), y=y_std_sk.flatten(), ax=ax, color='tab:brown', label='Datos estandarizados (SK-Learn)')
plt.xlabel('$x_1$')
plt.ylabel('y')
plt.show()
```

##### Comparando ambos

```python
fig, ax = plt.subplots(1,2, figsize=(15,7))

sns.scatterplot(x=X_norm.flatten(), y=y_norm.flatten(), ax=ax[0], color='tab:red', label='Implementacion propia')
sns.scatterplot(x=X_norm_sk.flatten(), y=y_norm_sk.flatten(), ax=ax[1], color='purple', label='MinMaxScaler (Sk-Learn)')

fig.suptitle('Comparativa implementacion de Normalizacion')

ax[0].set_xlabel('$x_1$')
ax[0].set_ylabel('y')
ax[1].set_xlabel('$x_1$')
ax[1].set_ylabel('y')

plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize=(15,7))

sns.scatterplot(x=X_std.flatten(), y=y_std.flatten(), ax=ax[0], color='tab:green', label='Implementacion propia')
sns.scatterplot(x=X_std_sk.flatten(), y=y_std_sk.flatten(), ax=ax[1], color='tab:brown', label='StandarScaler (Sk-Learn)')

fig.suptitle('Comparativa implementacion de Estandarizacion')

ax[0].set_xlabel('$x_1$')
ax[0].set_ylabel('y')
ax[1].set_xlabel('$x_1$')
ax[1].set_ylabel('y')

plt.show()
```

## Estructuras de Datos: Tuplas, Conjuntos y Diccionarios

## 1. Tuplas

**Concepto:** Son como listas pero no se pueden modificar después de crearlas (inmutables). Guardan datos ordenados.

**Declaración:**

```python
# Se declaran con paréntesis
mi_tupla = (1, 2, 3, 4, 5)
frutas = ("manzana", "naranja", "pera")
mixta = (1, "hola", 3.14)

print(mi_tupla)
print(frutas)
```

**Acceso a elementos:**

```python
# Se accede por índice (empieza en 0)
colores = ("rojo", "verde", "azul", "amarillo")

print("Primer elemento:", colores[0])
print("Último elemento:", colores[-1])
print("Del 1 al 3:", colores[1:3])
```

**Uso en ciclos:**

```python
numeros = (10, 20, 30, 40)

for num in numeros:
    print(num)
```

**Métodos principales:**

```python
tupla = (1, 2, 3, 2, 4, 2)

# count() - contar cuántas veces aparece un elemento
print("Veces que aparece el 2:", tupla.count(2))

# index() - encontrar la posición de un elemento
print("Posición del 3:", tupla.index(3))

# len() - tamaño de la tupla
print("Tamaño:", len(tupla))
```

## 2. Conjuntos (Sets)

**Concepto:** Colección de elementos únicos (sin repetidos) y sin orden específico.

**Declaración:**

```python
# Se declaran con llaves {}
conjunto = {1, 2, 3, 4, 5}
letras = {'a', 'b', 'c'}

# Los duplicados se eliminan automáticamente
con_duplicados = {1, 2, 2, 3, 3, 4}
print(con_duplicados)  # {1, 2, 3, 4}
```

**Acceso a elementos:**

```python
# No se puede acceder por índice, solo verificar si existe
colores = {"rojo", "verde", "azul"}

print("rojo" in colores)  # True
print("negro" in colores)  # False
```

**Uso en ciclos:**

```python
numeros = {1, 2, 3, 4, 5}

for num in numeros:
    print(num)
```

**Métodos principales:**

```python
conjunto = {1, 2, 3}

# add() - agregar elemento
conjunto.add(4)
print(conjunto)

# remove() - eliminar elemento
conjunto.remove(2)
print(conjunto)

# Operaciones entre conjuntos
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print("Unión:", a | b)           # {1, 2, 3, 4, 5, 6}
print("Intersección:", a & b)     # {3, 4}
print("Diferencia:", a - b)       # {1, 2}
```

## 3. Diccionarios

**Concepto:** Guardan pares de clave-valor. Sirven para asociar datos relacionados.

**Declaración:**

```python
# Se declaran con llaves y dos puntos
persona = {
    "nombre": "Juan",
    "edad": 25,
    "ciudad": "CDMX"
}

calificaciones = {"mate": 90, "física": 85, "química": 88}

print(persona)
print(calificaciones)
```

**Acceso a elementos:**

```python
estudiante = {"nombre": "María", "edad": 22, "carrera": "Ingeniería"}

# Acceder con corchetes
print(estudiante["nombre"])

# Acceder con get() (más seguro)
print(estudiante.get("edad"))
print(estudiante.get("promedio", "No existe"))  # valor por defecto

# Modificar o agregar
estudiante["edad"] = 23
estudiante["semestre"] = 5
print(estudiante)
```

**Uso en ciclos:**

```python
notas = {"mate": 95, "física": 88, "química": 92}

# Iterar sobre claves
for materia in notas:
    print(materia)

# Iterar sobre valores
for nota in notas.values():
    print(nota)

# Iterar sobre clave y valor
for materia, nota in notas.items():
    print(f"{materia}: {nota}")
```

**Métodos principales:**

```python
d = {"a": 1, "b": 2, "c": 3}

# keys() - obtener claves
print("Claves:", list(d.keys()))

# values() - obtener valores
print("Valores:", list(d.values()))

# items() - obtener pares
print("Items:", list(d.items()))

# pop() - eliminar y retornar
valor = d.pop("b")
print("Eliminado:", valor)
print("Diccionario:", d)

# update() - actualizar con otro diccionario
d.update({"d": 4, "e": 5})
print("Actualizado:", d)
```
